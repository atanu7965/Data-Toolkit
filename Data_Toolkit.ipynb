1. What is NumPy, and why is it widely used in Python ?
   - The core object in NumPy is the ndarray (N-dimensional array), which is a grid of homogeneous (same-type) data elements. Unlike Python's built-in lists, which can store elements of different data types and are less efficient for numerical tasks, all elements of a NumPy array must be of the same type (e.g., all integers or all floating-point numbers).
   NumPy's popularity stems primarily from its ability to bridge Python's ease of use with the high performance required for computationally intensive tasks.

Superior Performance: NumPy's core routines are implemented in highly optimized C and Fortran, allowing it to perform mathematical operations on large datasets much faster than native Python lists. Operations can be up to 50x faster.

Memory Efficiency: NumPy arrays store data in a contiguous block of memory, which allows for efficient storage and access, especially when dealing with large datasets. This compact representation lets NumPy fit more data into a computer's fast RAM.

Vectorized Operations & Broadcasting: NumPy enables vectorization, meaning mathematical and logical operations are applied to entire arrays at once, rather than requiring explicit loops in Python code. The broadcasting feature further simplifies operations on arrays of different shapes by automatically adjusting their dimensions, leading to cleaner and more readable code.

2. How does broadcasting work in NumPy?
   - NumPy uses a specific set of rules to determine if arrays can be broadcast together, comparing dimensions from right to left.
Arrays with fewer dimensions are padded with ones on the left.
For broadcasting to be possible, the dimension sizes must either be equal or one of them must be 1.
If dimension sizes are incompatible and neither is 1, a ValueError occurs.
The resulting array will have the highest number of dimensions from the input arrays, with each dimension's size being the maximum of the corresponding input dimensions.

6. What are the differences between NumPy arrays and Python lists?
    - 1. Data Type

Python List: Can store elements of different data types (heterogeneous).

NumPy Array: Stores elements of the same data type (homogeneous).

2. Performance

Python List: Slower for numerical computations because operations are performed element by element.

NumPy Array: Much faster for numerical operations due to optimized C implementation and vectorization.

3. Memory Usage

Python List: Uses more memory because it stores references to objects.

NumPy Array: Uses less memory as it stores data in a contiguous memory block.

4. Mathematical Operations

Python List: Does not support direct mathematical operations (e.g., list1 + list2 concatenates lists).

NumPy Array: Supports element-wise mathematical operations (addition, subtraction, multiplication, etc.).

5. Multidimensional Support

Python List: Can represent multidimensional data using nested lists, but handling is complex.

NumPy Array: Built-in support for multidimensional arrays (1D, 2D, 3D, etc.) with easier indexing.

6. Functionality

Python List: Provides general-purpose methods like append(), remove(), pop().

NumPy Array: Provides a large collection of mathematical, statistical, and linear algebra functions.

7. Size Modification

Python List: Dynamic in size (can grow or shrink easily).

NumPy Array: Fixed size after creation (resizing creates a new array).

8. Speed of Execution

Python List: Slower execution for large datasets.

NumPy Array: Faster execution due to vectorized operations.

9. Usage

Python List: Best for general-purpose programming and storing mixed data.

NumPy Array: Best for scientific computing, data analysis, machine learning, and numerical tasks.

8. What does the term “vectorized operation” mean in NumPy?

Missing values can distort graphs and charts. Cleaning data ensures accurate visual representation.

7. Maintains Data Integrity

Handling missing data ensures relationships between columns and rows remain meaningful.

15. How does NumPy handle multidimensional arrays ?
    - NumPy handles multidimensional arrays efficiently through its ndarray (N-dimensional array) object, which is designed for fast numerical computation. The key points are:

1. N-Dimensional Array (ndarray)

NumPy provides the ndarray, which can represent arrays of any dimension:

1D array → Vector

2D array → Matrix

3D or higher → Tensors

Each array has attributes like ndim (number of dimensions), shape, and size.

2. Shape and Axes

The shape of an array defines the size along each dimension (e.g., (3, 4) for 3 rows and 4 columns).

Each dimension is called an axis:

Axis 0 → Rows

Axis 1 → Columns

3. Efficient Memory Layout

Multidimensional arrays are stored in contiguous memory blocks.

This allows faster access and computation compared to nested Python lists.

4. Indexing and Slicing

NumPy allows easy access to elements using multi-indexing:

a[i, j] for 2D arrays

Supports slicing, boolean indexing, and fancy indexing for multidimensional data.

5. Vectorized Operations

Operations are applied element-wise across the entire multidimensional array without explicit loops.

Example: adding two 2D arrays directly.

6. Broadcasting

NumPy can perform operations between arrays of different shapes using broadcasting.

Smaller arrays are automatically expanded to match larger ones when possible.

7. Built-in Functions

NumPy provides functions like reshape(), transpose(), flatten(), sum(axis=), and mean(axis=) to manipulate multidimensional arrays easily.

17. Explain the difference between apply() and map() in Pandas.
    - In Pandas, apply() and map() are used to apply functions to data, but they differ in scope, flexibility, and usage.

1. Scope of Operation

map()

Works only on a Series (single column).

Applies a function element-wise to each value in the Series.

apply()

Works on Series and DataFrame.

Can apply a function row-wise or column-wise on a DataFrame.

2. Type of Function

map()

Mainly used with:

A function

A dictionary

A Series (for value mapping)

apply()

Used with:

Custom functions

Lambda functions

More complex logic involving multiple values

3. Axis Parameter

map()

No axis parameter (operates on individual elements only).

apply()

Uses axis:

axis=0 → apply to each column

axis=1 → apply to each row

4. Return Type

map()

Returns a Series.

apply()

Returns a Series or DataFrame, depending on the function output.

5. Performance

map()

Faster for simple element-wise operations on a single column.

apply()

Slower than map() for simple tasks but more powerful and flexible.

6. Use Cases

map()

Replacing values using a dictionary

Applying simple transformations to one column

apply()

Performing calculations across rows or columns

Creating new columns based on multiple existing columns

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd

data = {
    'Marks': [60, 70, 80, 90, 100],
    'Age': [18, 19, 20, 21, 22]
}

df = pd.DataFrame(data)

mean_marks = df['Marks'].mean()

print("Mean of Marks column:", mean_marks)
```

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Maths': [60, 70, 80, 90, 100],
    'Physics': [65, 75, 85, 95, 105],
    'Chemistry': [58, 68, 78, 88, 98]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

print("Correlation Matrix:")
print(corr_matrix)

sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.
import pandas as pd

data = {
    'Marks': [60, 70, 80, 90, 100],
    'Age': [18, 19, 20, 21, 22]
}

df = pd.DataFrame(data)

df['Grade'] = df['Marks'].apply(lambda x: 'A' if x >= 85 else 'B')

print(df)
```

```python
#8. Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]   # Line 1
y2 = [1, 3, 5, 7, 9]    # Line 2
y3 = [5, 3, 6, 2, 7]    # Line 3

plt.plot(x, y1, marker='o', label='Line 1')
plt.plot(x, y2, marker='s', label='Line 2')
plt.plot(x, y3, marker='^', label='Line 3')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()

plt.show()
```

```python
#9.  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Marks': [60, 75, 90, 85, 50]
}

df = pd.DataFrame(data)

threshold = 70

filtered_df = df[df['Marks'] > threshold]

print("Filtered DataFrame (Marks > 70):")
print(filtered_df)
```

```python
#11. Perform matrix multiplication using NumPy.
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

C = np.dot(A, B)

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nMatrix Multiplication (A x B):")
print(C)
```
